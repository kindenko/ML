```python
x = train_df.values[:,  0]
label_encoder = LabelEncoder().fit(train_df.values[:, 1])
y = label_encoder.transform(train_df.values[:, 1])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=SEED)
```

```python
TOKEN_PATTERN = '(?u)\\b\\w\\w+\\b'
```

```python
def tokenize(text):
    return re.findall(TOKEN_PATTERN, text.lower())
```

```python
#docs = [tokenize(text) for text in x_train]
```

```python
def lemmatize(token):
    lemmatizer_cache = {}
    if lemmatizer.word_is_known(token):
        if token not in lemmatizer_cache:
            lemmatizer_cache[token] = lemmatizer.parse(token)[0].normal_form
        return lemmatizer_cache[token]
    return token
```

# №1

```python
pipe = Pipeline([
    ('vectorizer', TfidfVectorizer(tokenizer=tokenize, lowercase=True, max_df=0.2, stop_words = stopword_set ,
                                   ngram_range=(1,2), min_df=2,
                                   token_pattern=("(?u)\\b\\w\\w+\\b"),
                                   preprocessor = lemmatize,
                                   norm='l2')),
    ('model', SGDClassifier(random_state=SEED, loss='log', class_weight='balanced', 
                            verbose=1, alpha = 0.000000001,
                             n_jobs=-1))  ])
```

# Лучшая модель + Bagging

```python
n_estimators = 5
answer = np.zeros((n_estimators, x_train.shape[0]), dtype=int)
```

```python
for i in tqdm_notebook(range(n_estimators)):
    DT = pipe
    index = np.random.choice(np.arange(x_train.shape[0]), x_train.shape[0])
    DT.fit(x_train[index], y_train[index])
    answer[i] = DT.predict(x_train)
```

```python
bagging_dt_answer = [np.bincount(answer[:, i]).argmax() for i in range(answer.shape[1])]
```

```python
balanced_accuracy_score(y_train, bagging_dt_answer)
```
